```python
import ast 
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df ['genres']= movies_df['genres'].apply(convert)
movies_df['keywords']= movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
import pandas as pd

data = {
    'user': ['A','A','A','B','B','C','C','D','D'],
    'movie': ['Iron Man','Avengers','Thor','Iron Man','Thor','Avengers','Thor','Iron Man','Avengers'],
    'rating': [5,4,4,5,3,5,4,4,5]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
pivot = df.pivot_table(index='user', columns='movie', values='rating')
pivot.fillna(0, inplace=True)

pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(pivot)
```

```python
def recommend_user(user):
    user_index = pivot.index.get_loc(user)
    distances = similarity[user_index]
    similar_users = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:3]

    for i in similar_users:
        print("Similar user:", pivot.index[i[0]])
```

```python
recommend_user('A')
```

```python
def recommend_movies(user):
    user_index = pivot.index.get_loc(user)
    distances = similarity[user_index]
    
    similar_users = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:3]
    
    recommended_movies = set()
    
    for i in similar_users:
        similar_user = pivot.index[i[0]]
        movies_watched = pivot.loc[similar_user]
        
        for movie, rating in movies_watched.items():
            if rating > 0 and pivot.loc[user][movie] == 0:
                recommended_movies.add(movie)
    
    print("Recommended Movies:")
    for movie in recommended_movies:
        print(movie)
```

```python
recommend_movies('A')
```
